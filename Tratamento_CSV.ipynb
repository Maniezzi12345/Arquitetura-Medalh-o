```python
df_tratado_csv = spark.read.format("csv").option("header","true").load("Files/paises.csv/Europa_informa√ß√µes.csv")

display(df_tratado_csv)
```

```python
df_tratado_csv = df_tratado_csv.drop("pib_usd")

display(df_tratado_csv)

```

```python
df_tratado_csv = df_tratado_csv.drop("pib_usd_padrao")
df_tratado_csv = df_tratado_csv.drop("pib_usd_milhoes")
display(df_tratado_csv)

```

```python
from pyspark.sql.functions import round, col
tipos = {
    "iso_code":"string",
    "pib_per_capita": "double",
    "idh": "double",
    "co2_per_capita":"double",
    "expectativa_vida": "double",
    "energia_renovavel_pct": "double",
    "internet_pc": "double",
    "desemprego":"double"
}

for coluna, tipo in tipos.items():
    df_tratado_csv = df_tratado_csv.withColumn(coluna, col(coluna).cast(tipo))

display(df_tratado_csv)
```

```python
from pyspark.sql.functions import round, col

df_tratado_csv = df_tratado_csv.fillna({
    'iso_code': 'N√£o informado', 
    'pib_per_capita': 0,
    'idh': 0,
    'desemprego': 0,
    'expectativa_vida':0,
    'co2_per_capita':0,
    'energia_renovavel_pct':0,
    'internet_pc':0,
    'desemprego':0})

df_tratado_csv = df_tratado_csv.dropDuplicates(["iso_code"])

display(df_tratado_csv)
```

```python
from pyspark.sql.functions import col, trim, upper

df_tratado_csv = df_tratado_csv.withColumn("iso_code", trim(upper(col("iso_code"))))
df_tratado_csv = df_tratado_csv.withColumnRenamed("iso_code", "codigo_iso_2") 
display(df_tratado_csv)

```

```python

df_tratado_csv = df_tratado_csv.filter((df_tratado_csv.idh >= 0) & (df_tratado_csv.idh <= 1))
df_tratado_csv = df_tratado_csv.filter((df_tratado_csv.desemprego >= 0) & (df_tratado_csv.desemprego <= 100))
df_tratado_csv = df_tratado_csv.filter((df_tratado_csv.expectativa_vida >= 0) & (df_tratado_csv.expectativa_vida <= 90))
df_tratado_csv = df_tratado_csv.filter((df_tratado_csv.co2_per_capita >= 0) & (df_tratado_csv.co2_per_capita <= 50))
df_tratado_csv = df_tratado_csv.filter((df_tratado_csv.energia_renovavel_pct >= 0) & (df_tratado_csv.energia_renovavel_pct <= 100))
df_tratado_csv = df_tratado_csv.filter((df_tratado_csv.internet_pc >= 0) & (df_tratado_csv.internet_pc <= 100))
df_tratado_csv = df_tratado_csv.filter((df_tratado_csv.pib_per_capita >= 0) & (df_tratado_csv.pib_per_capita <= 200000))
display(df_tratado_csv)
```
